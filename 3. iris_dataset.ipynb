## pre-processing

## Linear transformation

### standard scaler technique (-1 to +1)

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

### fit and transform methods : proper scaling of the data

```python
print(scaler.fit(X))
print(scaler.transform(X))         #outliars also possible
```

### MinMax scaler technique (0 to +1 range)

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
print(scaler.fit_transform(X))
```

### MaxAbsoluteScaler technique

```python
from sklearn.preprocessing import MaxAbsScaler
X,y = load_iris(return_X_y=True)
scaler = MaxAbsScaler()
print(scaler.fit_transform(X))
```

standard scalar
- z = [(x-x.mean)/std_deviation]

MinMaxScalar
- x = [(value-min)/(max-min)]

# Non Linear Transformation

### Quantile transformation

```python
from sklearn.preprocessing import QuantileTransformer
Quantile_Transformer = QuantileTransformer(random_state=0)        #--> state value is the seed value (the value we provided in the beginning to the model)
print(Quantile_Transformer.fit_transform(X))
```

```python
from sklearn.preprocessing import QuantileTransformer
Quantile_Transformer = QuantileTransformer()        
print(Quantile_Transformer.fit_transform(X))
```

```python
from sklearn.preprocessing import QuantileTransformer
Quantile_Transformer = QuantileTransformer(random_state=0)        
print(Quantile_Transformer.fit_transform(X))
```

splitting the original dataset into two parts - some X (and y corresponding to these X)for training and some for testing


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)   #out of 150 rows, 30%(0.30) is for testing and rest for training(70%)
print(X_train.shape, y_train.shape)                                                          #this random_state values chooses these rows randomly, that's why 0 is assigned
print(X_test.shape, y_test.shape)                                                            #while testing, it has to take the same rows for testing and training. so to restrict the model to this condition, we set the random_state. so it can be any integer value
                                                                                            #same random selection everytime
```

```python
from sklearn.preprocessing import QuantileTransformer
Quantile_Transformer = QuantileTransformer(random_state=0)         #this random_state value has to be the same as the above random_state value (so zero)
print(Quantile_Transformer.fit_transform(X_train))
```
