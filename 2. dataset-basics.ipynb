###splitting dataset

```python
# two ways 
# 1. train test split
# 2. kfold splitting
```

```python
from sklearn.datasets import load_diabetes
import pandas as pd
x, y = load_diabetes(return_X_y=True)
x=pd.DataFrame(x)
y=pd.DataFrame(y)
print(x)
print(y)
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=5, random_state=42, shuffle=False) 
print(x_train)
print(y_train) 
print(x_test) 
print(y_test) 
```

```python
#k-fold splitting

from sklearn.model_selection import KFold
i= 1
kf = KFold(n_splits=10)
for train_set, test_set in kf.split(x):
    print("iteration:", i)
    print(train_set, "having length:", len(train_set))
    print(test_set, "having length:", len(test_set))
    print("----------------------------------------------------------------------------------------------------")
    i+=1
```
