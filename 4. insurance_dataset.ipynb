### One-hot-encoder method for data- preprocessing

converting values to numeric value for preprocessing
e.g. - smoker is either yes or no. so there will be two diff variables and 0 and 1 will be represented depending on whether the person is a non-smoker or smoker

```python
#taking categorical columns - sex, smoker and region
categorical_values = ['sex',  'smoker', 'region']
df_encoder = pd.get_dummies(data=df, prefix='OHE', prefix_sep='_', columns=categorical_values, drop_first=True, dtype='int8')   #drop_first drops the index
df_encoder                                                                                                                       #data type = integer - 8 bytes (int8)

#OHE_yes is for smoker. if 1 => smoker. if 0 => non-smoker
```

```python
x = df_encoder.drop(['charges'], axis=1)   #x contains all the features. dropping the 'charges' column - axis=1 specifies that we want to drop the entire column
y= df_encoder['charges']   #y contains the target
print(x.shape, y.shape)
```

### including children for OHE too

```python
categorical_values = ['sex',  'smoker', 'region', 'children']
df_encoder = pd.get_dummies(data=df, prefix='OHE', prefix_sep='_', columns=categorical_values, drop_first=True, dtype='int8')   
df_encoder        
```

```python
x = df_encoder.drop(['charges'], axis=1) 
y= df_encoder['charges']   
print(x.shape, y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_new = scaler.fit_transform(x)
print(x_new)
print(x_new.shape)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=4)          #selects the most relevant features only
pc = pca.fit_transform(x_new)
pc
```

### training and testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(pc, y, test_size=0.30, random_state=0)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
model_DT_Reg = DecisionTreeRegressor()
model_DT_Reg.fit(x_train, y_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
y_pred = model_DT_Reg.predict(x_test)
y_pred
```

```python
df = pd.DataFrame({'Real Values':y_test, 'Predicted Values':y_pred})
df
```

```python
#score of the model
model_score = model_DT_Reg.score(x_test, y_test)
model_score
```
