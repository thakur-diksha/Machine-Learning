```python
import pandas as pd
df=pd.read_csv('Wholesale customers data.csv')
```

```python
df
```

```python
 from sklearn.preprocessing import normalize
 data=normalize(df)
 data=pd.DataFrame(data,columns=df.columns)
 data
```

```python
#use dendrogram to know number of clusters
#take a line from y-axis parallel to x-axis in such a way that it crosses the dendrogram min num of times...that will give n_clusters value
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,10))
plt.title('Dendrogram')
dendo=shc.dendrogram(shc.linkage(data,method='ward'))
plt.axhline(y=6,color='r',linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
```

```python
cluster.fit_predict(data)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data['Milk'],data['Grocery'],c=cluster.labels_)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data['Fresh'],data['Grocery'],c=cluster.labels_)
```
