##feature-scaling

### 4 types -:
- Standard Scaler
- Min max
- Quantile 
- Absolute max

```python
#STANDARD SCALER

from sklearn.preprocessing import StandardScaler
arr = [[31, 67, 98], [54, 99, 1900], [1, -999, 765]]
sc = StandardScaler()
res = sc.fit_transform(arr)
print(sc.mean_)
print(sc.scale_)
print(res)
```

```python
#MIN MAX

from sklearn.preprocessing import MinMaxScaler
arr = [[31, 67, 98], [54, 99, 1900], [1, -999, 765]]
sc = MinMaxScaler(feature_range=[3, 10])
res = sc.fit_transform(arr)
# print(sc.mean_)
# print(sc.scale_)
print(res)
```

```python
#QUANTILE

from sklearn.preprocessing import QuantileTransformer
arr = [[31, 67, 98], [54, 99, 1900], [1, -999, 765]]
sc = QuantileTransformer()
res = sc.fit_transform(arr)
# print(sc.mean_)
# print(sc.scale_)
print(res)
```

```python
#MAX ABSOLUTE 

from sklearn.preprocessing import MaxAbsScaler
arr = [[31, 67, 98], [54, 99, 1900], [1, -999, 765]]
sc = MaxAbsScaler()
res = sc.fit_transform(arr)
# print(sc.mean_)
# print(sc.scale_)
print(res)
```

```python
# 2. replace it with mean

import numpy as np
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values=np.NaN, strategy='constant', fill_value=100)
si.fit_transform([[1,2,3],[4,np.NaN,6],[10,5,9]])
```
